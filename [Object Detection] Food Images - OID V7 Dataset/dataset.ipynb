## Downloading images

```python
def convert_and_save_annotations(df, folder_path):
    os.makedirs(folder_path, exist_ok=True)
    # Annotations for each image
    annotations_dict = {}

    for _, row in df.iterrows():
        # Extracting bounding box coordinates
        min_x, max_x, min_y, max_y = row["XMin"], row["XMax"], row["YMin"], row["YMax"]
        
        # Calculating center and size of the bounding box
        bbox_center_x = (min_x + max_x) / 2
        bbox_center_y = (min_y + max_y) / 2
        bbox_width = max_x - min_x
        bbox_height = max_y - min_y
        
        # Label entry with YOLO format
        image_id = row["ImageID"]
        label_entry = f"0 {bbox_center_x:.4f} {bbox_center_y:.4f} {bbox_width:.4f} {bbox_height:.4f}"
        
        if image_id not in annotations_dict:
            annotations_dict[image_id] = label_entry
        else:
            annotations_dict[image_id] += f"\n{label_entry}"

    # Save each label to a .txt file with the image ID as the filename
    for img_id, annotation_text in annotations_dict.items():
        label_file_path = os.path.join(folder_path, f"{img_id}.txt")
        with open(label_file_path, "w") as label_file:
            label_file.write(annotation_text)
```

```python
convert_and_save_annotations(df_train_ann, "labels/train")
convert_and_save_annotations(df_validation_ann, "labels/validation")
```

- Saving training and validation images IDs

```python
with open("val_img_ids.txt", "w") as val_file:
    val_ids = ["validation/" + img_id for img_id in df_validation_ann["ImageID"].unique()]
    val_file.write("\n".join(val_ids))

with open("train_img_ids.txt", "w") as train_file:
    train_ids = ["train/" + img_id for img_id in df_train_ann["ImageID"].unique()]
    train_file.write("\n".join(train_ids))
```
